```python
# Prepare L2-L3 requirements trace data for SQLite import

import pandas as pd
df = pd.read_csv("../data_sources/requirements_traces_raw.csv")
df.head(10)
```

```python
# Split L3 requirements into individual rows and remove empty traces

df['l3_requirement_id'] = df['l3_requirement_id'].str.split(',')
df = df.explode('l3_requirement_id')
df['l3_requirement_id'] = df['l3_requirement_id'].str.strip()
df = df.dropna(subset=['l3_requirement_id'])
df[['l2_requirement_id', 'l3_requirement_id']]
```

```python
# Reset index after exploding rows

df = df.reset_index(drop=True)
df.head(20)
```

```python
# Remove duplicate L2-L3 trace relationships after exploding comma-separated values

df = df.drop_duplicates(subset=['l2_requirement_id', 'l3_requirement_id'])
```

```python
# Export cleaned file

df.to_csv("../outputs_clean/requirements_traces_clean.csv", index=False)
```
