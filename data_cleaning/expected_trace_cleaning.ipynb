```python
# Prepare L2 requirement allocation data for SQLite import

import pandas as pd
df = pd.read_csv("../data_sources/l2_requirements_allocations_raw.csv")
df.head(10)
```

```python
# Allocated subsystems are separated onto individual rows

df['allocated_subsystem'] = df['allocated_subsystem'].str.split(',')
df = df.explode('allocated_subsystem')
df['allocated_subsystem'] = df['allocated_subsystem'].str.strip()
df[['l2_requirement_id', 'allocated_subsystem']]
```

```python
# Assign l3_subsystem_id based on allocated subsystem and validate mappings

df['l3_subsystem_id'] = df['allocated_subsystem'].map({
    'subsystem1': 1,
    'subsystem2': 2
})
df[['allocated_subsystem', 'l3_subsystem_id']]
```

```python
# Check for unmapped values in l3_subsystem_id column

df['l3_subsystem_id'].isna().sum()
```

```python
# Select final output columns

df = df[['l2_requirement_id', 'l3_subsystem_id', 'allocated_subsystem']]
df.head(10)
```

```python
# Reset index after exploding rows

df = df.reset_index(drop=True)
df.head(20)
```

```python
# Export cleaned file

df.to_csv("../outputs_clean/l2_requirements_allocations_clean.csv", index=False)
```
