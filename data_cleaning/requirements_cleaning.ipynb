```python
# Prepare L2 and L3 requirements data for SQLite import

import pandas as pd
df = pd.read_csv("../data_sources/requirements_all_raw.csv")
df.head(10)
```

```python
# Convert blank values in l3_subsystem_id to nulls

df['l3_subsystem_id'] = df['l3_subsystem_id'].astype('string').str.strip()
df['l3_subsystem_id'] = df['l3_subsystem_id'].replace(r'^\s*$', pd.NA, regex=True)
```

```python
# Trim whitespace from l3_subsystem_id values

df['l3_subsystem_id'] = df['l3_subsystem_id'].astype('string').str.strip()
```

```python
# Validate l3_subsystem_id values

df['l3_subsystem_id'].value_counts(dropna=False)
```

```python
# Convert l3_subsystem_id from float-like values (1.0, 2.0) to clean integer

df['l3_subsystem_id'] = pd.to_numeric(df['l3_subsystem_id'], errors='coerce')
df['l3_subsystem_id'] = df['l3_subsystem_id'].astype('Int64')
df['l3_subsystem_id'] = df['l3_subsystem_id'].astype('string')
```

```python
# Validate l3_subsystem_id values after datatype cleanup

df['l3_subsystem_id'].value_counts(dropna=False)
```

```python
# Export cleaned file

df.to_csv("../outputs_clean/requirements_all_clean.csv", index=False)
```
